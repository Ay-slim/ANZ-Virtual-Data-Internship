```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
%matplotlib inline
```

```python
#import dataset
df = pd.read_excel('ANZ synthesised transaction dataset.xlsx')
```

```python
#extract salary transactions
df_pred = df[df['txn_description']=='PAY/SALARY']
```

```python
df_pred.reset_index(inplace = True, drop = True)
```

```python
df_pred
```

```python
#check all columns
df_pred.columns
```

```python
#view missing values
df_pred.isna().sum()
```

```python
#extract latitude and longitude values
lat = [float(df_pred.long_lat[j][7:]) for j in range(len(df_pred))]
long = [float(df_pred.long_lat[j][0:6]) for j in range(len(df_pred))]
df_pred['long'] = long
df_pred['lat'] = lat
df_pred
```

```python
#Extract mean weekly salary for each unique account
unique_df = pd.DataFrame(
    df_pred.groupby(['account', 'gender', 'age', 'long', 'lat'], as_index = False)['amount'].mean())
```

```python
#Extract the average account balance for each account as an additional feature
averageBalance = df.groupby('account')['balance'].mean()
```

```python
#merge the two data frames
aggregate_df = pd.merge(unique_df, averageBalance, on ='account', how = 'left')
```

```python
#calculate annual salary from weekly salary and rename the amount column
aggregate_df['amount']=aggregate_df['amount']*52
aggregate_df.rename({'amount':'salary'}, axis = 1, inplace = True)
aggregate_df
```

```python
plt.hist(aggregate_df.salary, bins = 50);
```

```python
sns.scatterplot(data = aggregate_df, x = 'age', y = 'salary')
plt.grid()
plt.ylabel('Annual Salary')
plt.title('Scatter plot of Salary against age')
```

```python
sns.boxplot(data = aggregate_df, x = 'gender', y = 'salary')
plt.ylabel('Annual Salary')
plt.title('BoxPlot of Annual Salary against Gender')
```

```python
sns.scatterplot(data = aggregate_df, x = 'balance', y = 'salary')
plt.grid()
plt.xlabel('Average Balance')
plt.title('Scatter plot of Salary and average account balance')
```

Use KMeans to cluster the location coordinates

```python
#Elbow curve to check the optimum number of clusters
K_clusters = range(1,10)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
score = [kmeans[i].fit(aggregate_df[['long', 'lat']]).score(aggregate_df[['long', 'lat']]) for i in range(len(kmeans))]
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
#create 3 location clusters with kmeans
kmeans = KMeans(n_clusters = 3, init ='k-means++')
kmeans.fit(aggregate_df[['long', 'lat']]) # Compute k-means clustering.
aggregate_df['location_cluster_label'] = kmeans.fit_predict(aggregate_df[['long', 'lat']])
```

```python
#visualize salary by location clusters
sns.boxplot(data = aggregate_df, x = 'location_cluster_label', y = 'salary')
```

```python
#Apply dummy variables to gender
aggregate_df.gender = pd.get_dummies(aggregate_df.gender, drop_first = True)
aggregate_df
```

```python
#drop columns that are not needed
aggregate_df.drop(['account', 'long', 'lat'], axis = 1, inplace = True)
aggregate_df
```

```python
#check correlation with salary
aggregate_df.corr().salary.sort_values(ascending = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(aggregate_df.drop('salary', axis = 1), 
                                                    aggregate_df['salary'], test_size = 0.25, random_state =1)
```

```python
X_train
```

```python
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
LR = LinearRegression()
LR.fit(X_train_scaled, y_train)
y_hat = LR.predict(X_test_scaled)
```

```python
#Create baseline prediction
y_baseline = y_train.mean()*np.ones(len(y_train))
print("baseline MSE= ", mean_squared_error(y_baseline, y_train))
```

```python
y_hat_train = LR.predict(X_train_scaled)
print("Train MSE= ", mean_squared_error(y_hat_train, y_train))
```

```python
print("Test MSE", mean_squared_error(y_hat, y_test))
```

```python
DTR = DecisionTreeRegressor()
```

```python
DTR.fit(X_train_scaled, y_train)
y_hat_train_DT = DTR.predict(X_train_scaled)
y_hat_test_DT = DTR.predict(X_test_scaled)
print("Train Dec Tree MSE= ", mean_squared_error(y_hat_train_DT, y_train))
print("Test Dec Tree MSE", mean_squared_error(y_hat_test_DT, y_test))
```
